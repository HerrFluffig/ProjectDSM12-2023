```python
import pandas as pd
```

```python
data = pd.read_csv('OSD.csv')




```

```python
import matplotlib.pyplot as plt

```

```python



age_bins = [0, 20, 30, 40, 50, 60, 70, 120]  
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71+'] 


data['Age_Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, right=False)


plt.figure(figsize=(10, 6))  


revenue_by_age_group = data.groupby('Age_Group')['Revenue_Total'].sum()


revenue_by_age_group.plot(kind='bar', color='skyblue')


plt.title('Einnahmen nach Altersgruppen')
plt.xlabel('Altersgruppen')
plt.ylabel('Einnahmen')


plt.tight_layout()
plt.show()
```

```python

total_revenue_by_gender = data.groupby('Gender')['Revenue_Total'].sum()


plt.figure(figsize=(8, 6))  


total_revenue_by_gender.plot(kind='bar', color=['blue', 'pink']) 


plt.title('Gesamteinnahmen nach Geschlecht')
plt.xlabel('Geschlecht')
plt.ylabel('Gesamteinnahmen')


plt.xticks(ticks=[0, 1], labels=['Männlich', 'Weiblich'])


plt.tight_layout()
plt.show()
```

```python
data['Purchase_DATE'] = pd.to_datetime(data['Purchase_DATE'], format='%d.%m.%y')


purchase_by_month = data.groupby(data['Purchase_DATE'].dt.to_period('M'))['N_Purchases'].sum()


plt.figure(figsize=(10, 6))  


purchase_by_month.plot(kind='bar', color='skyblue')


plt.title('Anzahl der transaktionen nach Monat')
plt.xlabel('Monat')
plt.ylabel('Anzahl der transaktionen')


plt.tight_layout()
plt.show()
```

```python
payment_counts = data['Pay_Method'].value_counts()


payment_labels = {
    0: 'Bargeld',
    1: 'Kreditkarte',
    2: 'Debitkarte',
    3: 'paypal'
}


payment_counts.index = payment_counts.index.map(payment_labels)


plt.figure(figsize=(8, 6))


payment_counts.plot(kind='bar', color='skyblue')


plt.title('Häufigste Zahlungsmethoden')
plt.xlabel('Zahlungsmethode')
plt.ylabel('Anzahl')


plt.tight_layout()
plt.show()
```
